```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
og_img = cv2.imread('00023.png',flags=cv2.IMREAD_GRAYSCALE)
plt.imshow(og_img,cmap='gray')
```

```python
og_img.shape
```

```python
BLACK = 0
WHITE = 255
```

```python
def get_upper_level_by_treshold(image,treshold):
    heigth,width = image.shape
    upper_level = np.zeros(shape=(heigth,width))
    for i in range(heigth):
        for j in range(width):
            if(image[i][j]>=treshold):
                upper_level[i][j]= WHITE
            else:
                upper_level[i][j]= BLACK
    return upper_level
```

```python
up=get_upper_level_by_treshold(og_img,50)
```

```python
plt.imshow(up,cmap='gray')
```

```python
fig = plt.figure(figsize=(250,300))
step = 25
for i,level in enumerate(range(255,0,-step)):
    up=get_upper_level_by_treshold(og_img,level)
    ax = fig.add_subplot(11,4,i+1)
    ax.imshow(up,cmap='gray')
plt.show()
```
